```python
from matplotlib import pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
from mpl_toolkits.axes_grid1 import make_axes_locatable
import IPython.display as ipd
import numpy as np
%matplotlib notebook
from scipy.signal import lfilter
```

```python
# parameter setting
amp = 1         # amplitude : 진폭 - 소리의 크기를 결정함
sr = 10000      # sampling rate(단위:Hz) : 신호가 1초에 몇번 들어가있나?
dur = 0.5       # seconds
freq = 100      # frequency(단위:Hz) : 1초에 한 사이클이 몇 번있나?

# sound is not discrete, but continuous. (연속적인 숫자를 컴퓨터는 쪼개서 처리함)
# 얼마나 쪼개냐? (sr) - 소리의 해상도가 달라짐.
# Hz : 1초에 몇번? 을 뜻하는 단위 - sr, freq 다른 개념이지만 Hz라는 단위를 씀
# freq : 1초에 반복되는 횟수 (고주파 : high frequency) -음의 높낮이를 결정
```

```python
# 소리는 아무리 복잡해도 sin wave의 형태
# 음의 높낮이 (freq), 음의 강도(amp)
# 두 변수가 독립적으로 작용함
```

```python
# generate time


t = np.arange(1, sr*dur+1)/sr

# 시간부터 만들어주어야 한다.
# 1/10000초부터 5000/10000초까지를 만들어줌 (+1한 것은 미만의 개념 때문)
# 시험문제 : lex(t) => 5000, 뒤에 +1이 없으면 4999
```

```python
# generate phase

theta = t*2*np.pi*freq

# 제일 중요한 단계
# sin은 time을 못 받아들이고, radiant값만을 받아들인다.
# 시간 자체가 각도와 상관없지만, freq 개념이 들어가면서 각도로의 변환이 가능해짐
# 한 freq는 무조건 2pi
# 시험문제 : len(t) = len(theta)
```

```python
# generate signal by cosine phasor
s=np.sin(theta)

# theta값은 radiant값이기 때문에 np.sin()에 넣어줄 수 있음
# len(t) = len(theta) = len(s)
```

```python
fig=plt.figure()                     # 이 함수를 물어보는 건 문제로 나오지 X
ax = fig.add_subplot(111)            # figure부터 만들어줌
ax.plot(t[0:1000],s[0:1000],'.')     # 이 부분이 중요.
ax.set_xlabel('time(S)')
ax.set_ylabel('real')
ipd.Audio(s, rate=sr)

# plotting할 때는 theta값을 쓰지 않고, time값과 sine값만 쓴다.
```

```python
c=np.exp(theta*1j)

# 여기서 j는 허수 i의 개념.
# 노트에 복소수에 대한 개념이 있음.
```

```python
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
ax.plot(t[0:1000], c.real[0:1000], c.imag[0:1000], '.')
ax.set_xlabel('time (s)')
ax.set_ylabel('real')
ax.set_zlabel('imag')
```
